```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import globals as gl
import os
from matplotlib.lines import Line2D

plt.style.use('default')
```

```python
pinfo = pd.read_csv(os.path.join(gl.baseDir, 'participants.tsv'), sep='\t')

data = pd.read_csv(os.path.join(gl.baseDir, f'merged_single_trial.tsv'), sep='\t')
data = data.groupby(['subNum', 'day', 'hand', 'chord', 'chordID']).mean(numeric_only=True).reset_index()

fig, axs = plt.subplots()

sns.lineplot(ax=axs, data=data[(data['trained_hand']==1)], x='day', y='trialPoint', hue='chord', errorbar='se', lw=2, palette=['darkgreen', 'darkmagenta'], err_kws={'lw': 0})
sns.lineplot(ax=axs, data=data[(data['trained_hand']==0)], x='day', y='trialPoint', hue='chord', errorbar='se', palette=['darkgreen', 'darkmagenta'], ls='--',err_kws={'lw': 0})

axs.set_xlabel('Day')
axs.set_ylabel('% success')

axs.set_title('Success rate for trained and untrained chords and hand')

# Get handles and labels
handles, labels = axs.get_legend_handles_labels()

filtered = [(h, l) for h, l in zip(handles, labels) if isinstance(h, Line2D)]

# Unpack filtered handles and labels
filtered_handles, filtered_labels = zip(*filtered)

# Set custom legend
axs.legend(filtered_handles, [
    'trained chords, trained hand',
    'untrained chords, trained hand',
    'trained chords, untrained hand',
    'untrained chords, untrained hand'
],)

plt.tight_layout()
```
