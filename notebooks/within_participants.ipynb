```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import globals as gl
import os

plt.style.use('default')
```

```python
sn = 100

pinfo = pd.read_csv(os.path.join(gl.baseDir, 'participants.tsv'), sep='\t')
group = pinfo[pinfo['subNum'] == sn].group.reset_index(drop=True)[0]

data = pd.read_csv(os.path.join(gl.baseDir, f'subj{sn}_single_trial.tsv'), sep='\t')

fig, axs = plt.subplots()

if group == 'L':
    sns.lineplot(ax=axs, data=data[(data['hand']==1) ], x='day', y='trialPoint', hue='chord')
elif group == 'R':
    sns.lineplot(ax=axs, data=data[(data['hand']==2)], x='day', y='trialPoint', hue='chord')
    
```
